```python
# Функция рандомной генерации даты
def gend():
    start_date = datetime(2015, 1, 23)
    end_date = datetime(2024, 1, 23)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y-%m-%d')

# Название продуктов и генерация по кол-ву товара и ценнику
PR = [
    ('Ручка', 'Концелярия', random.randint(1, 20000), random.randint(5, 150)),
    ('Часы', 'Аксессуары', random.randint(1, 50), random.randint(1000, 10000)),
    ('Карандаш', 'Концелярия', random.randint(1, 25000), random.randint(5, 50)),
    ('Брилок', 'Аксессуары', random.randint(1, 250), random.randint(20, 80)),
    ('Туфли', 'Одежда', random.randint(1, 1500), random.randint(2000, 3000)),
    ('Тетрадка', 'Концелярия', random.randint(1, 15000), random.randint(20, 200)),
    ('Рюкзак', 'Аксессуары', random.randint(1, 1500), random.randint(1500, 6000)),
    ('Мел', 'Концелярия', random.randint(1, 30000), random.randint(50, 450)),
    ('Линейка', 'Концелярия', random.randint(1, 17000), random.randint(50, 350)),
    ('Стирательная резинка', 'Концелярия', random.randint(1, 21000), random.randint(5, 25)),
    ('Очки', 'Аксессуары', random.randint(1, 150), random.randint(500, 2000)),
    ('Рубашка', 'Одежда', random.randint(1, 1800), random.randint(650, 1000)),
    ('Блокнот', 'Концелярия', random.randint(1, 12000), random.randint(50, 250)),
    ('Галстук', 'Одежда', random.randint(1, 2000), random.randint(300, 600)),
    ('Транспортир', 'Концелярия', random.randint(1, 14000), random.randint(100, 250)),
    ('Скрепка', 'Концелярия', random.randint(1, 40000), random.randint(2, 100)),
    ('Стол', 'Мебель', random.randint(1, 150), random.randint(3000, 10000)),
    ('Пиджак', 'Одежда', random.randint(1, 500), random.randint(2000, 6000)),
    ('Стул', 'Мебель', random.randint(1, 200), random.randint(2000, 7000))
]

# Запись в csv файл
with open('Report.csv', 'w', encoding='utf-8-sig', newline='') as file:
    writer = csv.writer(file, delimiter='|')

    # Заголовок в файле
    writer.writerow(['Номер заказа', 'Дата заказа', 'Название товара', 'Категория товара', 'Количество продаж', 'Цена за единицу', 'Общая стоимость'])
    viruch = 0
    
    # Запись данных
    for i in range(1, 20):
        rDT = gend()
        rPR = random.choice(PR)
        nPR = rPR[0]
        categ = rPR[1]
        coll = rPR[2]
        price = rPR[3]
        
        viruch += coll*price
        
        writer.writerow([random.randint(1, 100000000), rDT, nPR, categ, coll, price, coll*price])
        
```
